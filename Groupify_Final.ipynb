```python
#Multiplicative
colname='A'
maxratingoutofbond=6
mul=1
for i in range (0,len(group_users)):
    mul=mul*df[str(chr(ord(colname) + i))]
df['Multiplicative'] = mul
df.sort_values(by=['Multiplicative'], ascending=False).head(15)
```

```python
#Average without misery
colname='A'
sumrow=0
counter=0
cnvtlist=[]
for j in range (0,len(df)):
    counter=0
    sumrow=0
    for i in range (0,len(group_users)):
        pos=str(chr(ord(colname) + i))
        value = df.at[j,pos]
        if(value>1):
            sumrow=sumrow+value
            counter=counter+1
    if(counter>0):
        cnvtlist.append(sumrow/counter)
    else:
        cnvtlist.append(0)
ser = pd.Series(cnvtlist)
df['Average Without misery']=ser
df.sort_values(by=['Average Without misery'],ascending=False).head(1663)
```

```python
#Approval Voting
colname='A'
counter=0
newlist=[]
for j in range (0,len(df)):
    counter=0
    for i in range (0,len(group_users)):
        pos=str(chr(ord(colname) + i))
        value = df.at[j,pos]
        if(value>2):
            counter=counter+1
    if(counter>0):
        newlist.append(counter)
    else:
        newlist.append(0)
ser = pd.Series(newlist)
df['Approval Voting']=ser
df.sort_values(by=['Approval Voting'],ascending=False).head(10)
```

```python
#Copeland rule
colname='A'
newlist=[]
for j in range (0,len(df)):

    for i in range (0,len(group_users)):
        pos1=str(chr(ord(colname) + i))
        val1 = df.at[j,pos1]
        beats = 0
        lose = 0
        for k in range (0,len(group_users)):
           pos2=str(chr(ord(colname) + k))
           val2 = df.at[j,pos2]
           if(val1 > val2):
             beats = beats+1
           elif(val2 > val1) :
             lose = lose+1

        cope = beats - lose
        newlist.append(cope)

ser = pd.Series(newlist)
df['Copeland Rule']=ser
df.sort_values(by=['Copeland Rule'],ascending=False).head(16)
```

```python
#Bordacount
colname='A'
bor = 0
members = len(group_users)
for i in range (0,len(group_users)):
    bor += (members-i)*df[str(chr(ord(colname) + i))]
df['Borda'] = bor
df.sort_values(by=['Borda'], ascending=False).head(100)
```
